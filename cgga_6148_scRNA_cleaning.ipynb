## Data cleaning of scRNA-seq data

- transpose matrix
- save as csv to use for PCA

```python
## import packages for cleaning and EDA
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## load data and examine head of data
filename = "CGGA.scRNAseq_6148.count.matrix.tsv"
df = pd.read_csv(filename, sep='\t')
df.head()
```

```python
df_2 = df.transpose()
df_2.head()
```

```python
# assign row 1 as column names then delete row 1 (to avoid duplicate labels)
df_2.columns = df_2.iloc[0]
df_3 = df_2.iloc[1:]
df_3.head()
```

```python
# convert and save as csv

df_3.to_csv("cleaned_cgga_6148_scRNA.csv", index=False)
```
